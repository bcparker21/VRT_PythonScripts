```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import geopandas as gpd
import seaborn as sns
import numpy as np
import datetime
from shapely.geometry import Point
%matplotlib inline
```

```python
OTP_current=pd.read_csv(r"N:\Planning - New File Structure\Performance\OTP_1_28_2_8_19.csv")
OTP_current['DepartTime']=pd.to_datetime(OTP_current['DepartTime'])
OTP_current['ScheduledStopTime']=pd.to_datetime(OTP_current['ScheduledStopTime'])
OTP_current['ArriveTime']=pd.to_datetime(OTP_current['ArriveTime'])
OTP_current['TripDate']=pd.to_datetime(OTP_current['TripDate'])
OTP_current['ScheduledTripStart']=pd.to_datetime(OTP_current['ScheduledTripStart'])
OTP_current['ScheduledTripEnd']=pd.to_datetime(OTP_current['ScheduledTripEnd'])
```

```python
OTP_18=pd.read_pickle(r"\\VRIDE-FS2\Development\Planning - New File Structure\Performance\OTP_FY18.pkl")
OTP_18['DepartTime']=pd.to_datetime(OTP_18['DepartTime'])
OTP_18['ScheduledStopTime']=pd.to_datetime(OTP_18['ScheduledStopTime'])
OTP_18['ArriveTime']=pd.to_datetime(OTP_18['ArriveTime'])
OTP_18['TripDate']=pd.to_datetime(OTP_18['TripDate'])
OTP_18['ScheduledTripStart']=pd.to_datetime(OTP_18['ScheduledTripStart'])
OTP_18['ScheduledTripEnd']=pd.to_datetime(OTP_18['ScheduledTripEnd'])
```

```python
OTP_current.dropna(axis=0,inplace=True,subset=['DepartTime'])
OTP_current['Invalid']=OTP_current['DepartTime']<OTP_current['ArriveTime']
OTP_current=OTP_current[OTP_current.Invalid==False]
```

```python
OTP_18.dropna(axis=0,inplace=True,subset=['DepartTime'])
OTP_18['Invalid']=OTP_18['DepartTime']<OTP_18['ArriveTime']
OTP_18=OTP_18[OTP_18.Invalid==False]
```

```python
OTP_current['Start']=np.where(OTP_current['ScheduledTripStart']==OTP_current['ScheduledStopTime'],True,False)
OTP_current['End']=np.where(OTP_current['ScheduledTripEnd']==OTP_current['ScheduledStopTime'],True,False)
OTP_current['Terminal']=np.logical_or(OTP_current['Start'],OTP_current['End'])
```

```python
OTP_18['Start']=np.where(OTP_18['ScheduledTripStart']==OTP_18['ScheduledStopTime'],True,False)
OTP_18['End']=np.where(OTP_18['ScheduledTripEnd']==OTP_18['ScheduledStopTime'],True,False)
OTP_18['Terminal']=np.logical_or(OTP_18['Start'],OTP_18['End'])
```

```python
OTP_current['RelDepart']=OTP_current['DepartTime']-OTP_current['ScheduledStopTime']
OTP_current['RelArrive']=OTP_current['ArriveTime']-OTP_current['ScheduledStopTime']
```

```python
OTP_18['RelDepart']=OTP_18['DepartTime']-OTP_18['ScheduledStopTime']
OTP_18['RelArrive']=OTP_18['ArriveTime']-OTP_18['ScheduledStopTime']
```

```python
OTP_current['RelDepNum']=(pd.to_numeric(OTP_current['RelDepart'],downcast='integer'))/60000000000
OTP_current['RelArrNum']=(pd.to_numeric(OTP_current['RelArrive'],downcast='integer'))/60000000000
```

```python
OTP_18['RelDepNum']=(pd.to_numeric(OTP_18['RelDepart'],downcast='integer'))/60000000000
OTP_18['RelArrNum']=(pd.to_numeric(OTP_18['RelArrive'],downcast='integer'))/60000000000
```

```python
OTP_current['ArrStatus']='On Time'
OTP_current['ArrStatus'][OTP_current['RelArrNum']<-1]='Early'
OTP_current['ArrStatus'][OTP_current['RelArrNum']>5]='Late'
OTP_current['ArrStatus'][(OTP_current['Terminal'])&(OTP_current['RelArrNum']<5)]='On Time'
```

```python
OTP_18['ArrStatus']='On Time'
OTP_18['ArrStatus'][OTP_18['RelArrNum']<-1]='Early'
OTP_18['ArrStatus'][OTP_18['RelArrNum']>5]='Late'
OTP_18['ArrStatus'][(OTP_18['Terminal'])&(OTP_18['RelArrNum']<5)]='On Time'
```

```python
OTP_current['DepStatus']='On Time'
OTP_current['DepStatus'][OTP_current['RelDepNum']<-1]='Early'
OTP_current['DepStatus'][OTP_current['RelDepNum']>5]='Late'
```

```python
OTP_18['DepStatus']='On Time'
OTP_18['DepStatus'][OTP_18['RelDepNum']<-1]='Early'
OTP_18['DepStatus'][OTP_18['RelDepNum']>5]='Late'
```

```python
OTP_current['Status']='On Time'
OTP_current['Status'][OTP_current['DepStatus']=='Late']='Late'
OTP_current['Status'][OTP_current['DepStatus']=='Early']='Early'
OTP_current['Status'][(OTP_current['Terminal'])&(OTP_current['ArrStatus']=='OnTime')]='On Time'
```

```python
OTP_18['Status18']='On Time'
OTP_18['Status18'][OTP_18['DepStatus']=='Late']='Late'
OTP_18['Status18'][OTP_18['DepStatus']=='Early']='Early'
OTP_18['Status18'][(OTP_18['Terminal'])&(OTP_18['ArrStatus']=='OnTime')]='On Time'
```

```python
stopOTP_current=OTP_current.groupby(['StopName','Status'])
stopOTP_current=pd.DataFrame(stopOTP_current.size(),columns=['Count'])
stopOTP_current['Percent']=stopOTP_current.groupby(level=0).apply(lambda x: 100*x/x.sum())
stopOTP_current.reset_index(inplace=True)
```

```python
stopOTP_18=OTP_18.groupby(['StopName','Status18'])
stopOTP_18=pd.DataFrame(stopOTP_18.size(),columns=['Count18'])
stopOTP_18['Percent18']=stopOTP_18.groupby(level=0).apply(lambda x: 100*x/x.sum())
stopOTP_18.reset_index(inplace=True)
```

```python
gtfs_18_path=r"N:\Planning - New File Structure\GIS\Data\GTFS\VRT5_10_18\stops.txt"
gtfs_current_path=r"N:\Planning - New File Structure\GIS\Data\GTFS\VRT1_29_18\stops.txt"
```

```python
city_limits_path=r"N:\Planning - New File Structure\GIS\Data\Political\Boundaries\CityLimits\AllJurisdictions.shp"
```

```python
feed_current=pd.read_csv(gtfs_current_path)
feed_18=pd.read_csv(gtfs_18_path)
```

```python
stopOTP_current=stopOTP_current.merge(feed_current,left_on='StopName',right_on='stop_name')
stopOTP_current.head()
```

```python
stopOTP_18=stopOTP_18.merge(feed_18,left_on='StopName',right_on='stop_name')
stopOTP_18.head()
```

```python
city_limits=gpd.read_file("N:\\Planning - New File Structure\\GIS\\Data\\Political\\Boundaries\\CityLimits\\AllJurisdictions.shp")
city_limits=city_limits.to_crs(epsg=4326)
city_limits.crs
```

```python
stopOTP_current['Coordinates']=list(zip(stopOTP_current.stop_lon,stopOTP_current.stop_lat))
stopOTP_current['Coordinates']=stopOTP_current['Coordinates'].apply(Point)
stopOTP_current=gpd.GeoDataFrame(stopOTP_current,crs={'init':'epsg:4326','no_defs':True},geometry='Coordinates')
```

```python
stopOTP_18['Coordinates']=list(zip(stopOTP_18.stop_lon,stopOTP_18.stop_lat))
stopOTP_18['Coordinates']=stopOTP_18['Coordinates'].apply(Point)
stopOTP_18=gpd.GeoDataFrame(stopOTP_18,crs={'init':'epsg:4326','no_defs':True},geometry='Coordinates')
```

```python
stopOTP_current=gpd.sjoin(stopOTP_current,city_limits)
stopOTP_18=gpd.sjoin(stopOTP_18,city_limits)
```

```python
CityOTP_current=stopOTP_current.groupby(['CITY','Status']).sum()
CityOTP_current['Percent']=CityOTP_current.groupby(level=0).apply(lambda x: 100*x/x.sum())
```

```python
CityOTP_18=stopOTP_18.groupby(['CITY','Status18']).sum()
CityOTP_18['Percent18']=CityOTP_18.groupby(level=0).apply(lambda x: 100*x/x.sum())
```

```python
nampaStops_current=stopOTP_current.loc[stopOTP_current['CITY']=='NAMPA']
nampaStops_current=nampaStops_current.loc[nampaStops_current['Status']=='On Time']
```

```python
nampaStops_18=stopOTP_18.loc[stopOTP_18['CITY']=='NAMPA']
nampaStops_18=nampaStops_18.loc[nampaStops_18['Status18']=='On Time']
```

```python
keepCols18=['StopName','Status18','Count18','Percent18','stop_id','stop_lat','stop_lon','Coordinates']
nampaStops_18=nampaStops_18[keepCols18]
```

```python
keepCols_current=['StopName','Status','Count','Percent','stop_id','stop_lat','stop_lon','Coordinates']
nampaStops_current=nampaStops_current[keepCols_current]
```

```python
nampaStops=gpd.sjoin(nampaStops_current,nampaStops_18)
```

```python
nampaStops['Change']=nampaStops['Percent']-nampaStops['Percent18']
nampaStops
```
