```python
import pandas as pd
import geopandas as gpd
import gtfstk as gt
import numpy as np
from shapely.geometry import Point
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stopLevel=pd.read_excel(r"N:\Planning - New File Structure\Performance\FY18StopLevelRidership.xlsx",dtype={'StopId':np.str},sheet_name='Merge')
stopLevel.dropna(axis=0,how='all')
stopLevel.shape
```

```python
stopLevelTotal=stopLevel.groupby('StopId').sum()
stopLevelTotal.drop(['GpsLatitude','GpsLongitude'],inplace=True,axis=1)
stopLevelTotal.head()
```

```python
stopLevelCount=stopLevel.groupby('StopId').count()
stopLevelCount.drop(['GpsLatitude','GpsLongitude','SampleDate'],inplace=True,axis=1)
stopLevelCount.head()
```

```python
stopLevel=stopLevelTotal/stopLevelCount
boardingsByRoute=stopLevel.sum(axis=0)
boardingsByRoute
```

```python
stopLevelPct=stopLevel.divide(boardingsByRoute.T)
stopLevelPct.head(50)
```

```python
feed=gt.read_gtfs(r"N:\Planning - New File Structure\GIS\Data\GTFS\VRT5_10_18.zip",dist_units='mi')
stopLevel=stopLevel.merge(feed.stops,left_on='StopId',right_on='stop_id',how='outer')
stopLevel.head()
```

```python
crs={'init':'epsg:4326'}
```

```python
counties=gpd.read_file(r"N:\Planning - New File Structure\GIS\Data\Political\Boundaries\counties_id_ilso\counties_id_ilso.shp")
counties=counties.to_crs(epsg=4326)
# counties=counties.loc[counties['ABNAME']=='Ada County']|counties.loc[counties['ABNAME']=='Canyon County']
counties=counties[['ABNAME','geometry']]
counties.head()
```

```python
city_limits=gpd.read_file(r"N:\Planning - New File Structure\GIS\Data\Political\Boundaries\Census_City_Limits_All.shp")
city_limits=city_limits.to_crs(epsg=4326)
city_limits=city_limits[['NAME','geometry']]
city_limits=city_limits.append(counties)
```

```python
geometrySL=[Point(xy) for xy in zip(stopLevel.stop_lon,stopLevel.stop_lat)]
stopLevel=gpd.GeoDataFrame(stopLevel,crs=crs,geometry=geometrySL)
stopLevel=stopLevel.dropna(subset=['stop_lat','stop_lon'])
stopLevel=stopLevel.to_crs(epsg=4326)
stopLevel=gpd.sjoin(stopLevel,city_limits)
stopLevel.head()
```
